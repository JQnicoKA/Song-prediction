# Data Description

It is asked to predict the chances of a user listening to a song repetitively after the first observable listening event within a time window was triggered. If there are recurring listening event(s) triggered within a month after the user’s very first observable listening event, its target is marked 1, and 0 otherwise in the training set. The same rule applies to the testing set.

KKBOX provides a training data set consists of information of the first observable listening event for each unique user-song pair within a specific time duration. Metadata of each unique user and song pair is also provided. The use of public data to increase the level of accuracy of your prediction is encouraged.

The train and the test data are selected from users listening history in a given time period. Note that this time period is chosen to be before the WSDM-KKBox Churn Prediction time period. The train and test sets are split based on time, and the split of public/private are based on unique user/song pairs.

# Téléchargement du dataset

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import metrics, ensemble
````

# Préparation des donneées

````python
# Load data
df = pd.read_csv('train.csv')

# Load and join songs data
songs = pd.read_csv('songs.csv')
df = pd.merge(df, songs, on='song_id', how='left')
del songs

# Load and join songs data
members = pd.read_csv('members.csv')
df = pd.merge(df, members, on='msno', how='left')
del members
````

````python
df.info()
````

````python
# Count Na in %
df.isnull().sum()/df.isnull().count()*100
````

## Replace NA values and create dates
Set ```unknown``` value to categorical data and ```zero``` to quantitative variables

Extract year, month and day from the registration and expiration dates

````python
# Replace NA
for i in df.select_dtypes(include=['object']).columns:
    df.loc[df[i].isnull(), i]= 'unknown'
df = df.fillna(value=0)
````

````python
# registration_init_time
df.registration_init_time = pd.to_datetime(df.registration_init_time, format='%Y%m%d', errors='ignore')
df['registration_init_time_year'] = df['registration_init_time'].dt.year
df['registration_init_time_month'] = df['registration_init_time'].dt.month
df['registration_init_time_day'] = df['registration_init_time'].dt.day

# expiration_date
df.expiration_date = pd.to_datetime(df.expiration_date,  format='%Y%m%d', errors='ignore')
df['expiration_date_year'] = df['expiration_date'].dt.year
df['expiration_date_month'] = df['expiration_date'].dt.month
df['expiration_date_day'] = df['expiration_date'].dt.day
````

````python
df.head()
````

# Random Forest Tree

Let’s dive into how to create a random forest classifier.

A random forest is made up of decision trees. But that doesn’t mean that you need to actually create any decision trees : Scikit-Learn can handle this using the ```RandomForestClassifier``` class from the ```sklearn.ensemble``` module.

````python
# Dates to categoty
df['registration_init_time'] = df['registration_init_time'].astype('category')
df['expiration_date']        = df['expiration_date'].astype('category')

# Object data to category
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype('category')
    
# Encoding categorical features
for col in df.select_dtypes(include=['category']).columns:
    df[col] = df[col].cat.codes
````

````python
df.head()
````

Let’s understand what is doing the code below

1. create our features DataFrame ```X```, and our target Series ```y```
2. data is split using the ```train_test_split()``` function
3. instantiate the model ```model``` using the ```RandomForestClassifer```. We specified to create 250 different decision trees in our forest

````python
# Splitting the data and creating a model
X = df[df.columns[df.columns != 'target']]
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=100)

model = ensemble.RandomForestClassifier(n_estimators=100, max_depth=5)
````

Now it’s time to fit our data to the model. We can do this using the aptly-named ```.fit()``` method, which takes the training features and labels as inputs.

````python
model.fit(X_train, y_train)
````

Because we already have an array containing the true labels, we can easily compare the predictions to the ```y_test``` array. Scikit-learn comes with an ```accuracy_score()``` function that returns a ratio of accuracy.

````python
guessed = model.predict(X_test)

# Evaluating the model
print("Accuracy:",metrics.accuracy_score(y_test, guessed))
````

# Visualizing

One of the difficulties that you may run into in your machine learning journey is the black box of machine learning. Because libraries like Scikit-Learn make it so simple to create a random forest, it can be helpful to look at some of the details of your model.

````python
df_plot = pd.DataFrame({'features': df.columns[df.columns != 'target'],
                        'importances': model.feature_importances_})
df_plot = df_plot.sort_values('importances', ascending=False)
````

````python
plt.figure(figsize=[11,5])
sns.barplot(x = df_plot.importances, y = df_plot.features)
plt.title('Importances of Features Plot')
plt.show()
````

````python
from sklearn.tree import plot_tree

fig = plt.figure(figsize=(15, 10))
plot_tree(model.estimators_[0], 
          feature_names=list(X.columns),
          class_names=df['target'].unique().astype(str).tolist(), 
          filled=True, rounded=True,
          max_depth=4, fontsize=10)

plt.show()
````
